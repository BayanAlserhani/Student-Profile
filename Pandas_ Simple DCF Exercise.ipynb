```python
# Importing pandas as pd:

import pandas as pd
```

```python
# Parameters = Values:

starting_revenue = 222
Growth_factor = 1.18 
EBITDA_margin = 0.45 
Taxes_of_EBITDA = 0.21
Capital_margin = 0.03
Working_Capital_margin = 0.06
Discount_margin = 0.94
Exit_multiple = 32.70
Discount_Factor = 1
EBITDA_Exit_Multiple = 32.70
```

```python
# Calculations & Loop: 

years_num = pd.date_range(start='2022', end='2032', freq='Y')

Revenue_data = []
EBITDA_data = []
taxes_data = []
Capital_data = []
Working_Capital_data = []
Cashflow_data = []
Discount_data = []
PV_data = []

for i in range(1, len(years_num)+1):
    
    starting_revenue = starting_revenue * Growth_factor
    EBITDA = starting_revenue * EBITDA_margin
    taxes = EBITDA * Taxes_of_EBITDA
    Capital_Expenditures = starting_revenue * Capital_margin
    Working_Capital_Increase = starting_revenue * Working_Capital_margin
    cash_flow = EBITDA - taxes - Capital_Expenditures - Working_Capital_Increase
    Discount_Factor = Discount_Factor * Discount_margin
    Present_Value = cash_flow * Discount_Factor


    Revenue_data.append(starting_revenue)
    EBITDA_data.append(EBITDA)
    taxes_data.append(taxes)
    Capital_data.append(Capital_Expenditures)
    Working_Capital_data.append(Working_Capital_Increase)
    Cashflow_data.append(cash_flow)
    Discount_data.append(Discount_Factor)
    PV_data.append(Present_Value)
    
df_model = pd.DataFrame(index=years_num)
ter_df = pd.DataFrame()

    
```

```python
# Generating Revenue column:

df_model['Revenue'] = Revenue_data
df_model.round(2)
```

```python
# Generating EBITDA column:

df_model['EBITDA'] = EBITDA_data
df_model.round(2)
```

```python
# Generating Taxes column:

df_model['Taxes'] = taxes_data
df_model.round(2)
```

```python
# Generating Capital Expenditures column:

df_model['Capital Exp'] = Capital_data
df_model.round(2)
```

```python
# Generating Increase in Working Capital column:

df_model['Work Capital'] = Working_Capital_data
df_model.round(2)
```

```python
# Generating Cash Flow column:

df_model['Cash Flow'] = Cashflow_data
df_model.round(2)
```

```python
# Generating Discount Factor column:

df_model['Discount factor'] = Discount_data
df_model.round(2)
```

```python
# Generating Present Value of Future Cash Flow column:

df_model['Present Value FCF'] = PV_data
df_model.round(2)
```

```python
# Printing final statement:

print(f"Our projected Present Value of Projected Future Cash Flows are:", round(sum(PV_data),2),
      "\nWe are providing revenue guidance in 2032 to be:", round(Revenue_data[9]*Growth_factor, 2),
      "\nWith an estimated EBITDA around:",round(Revenue_data[9]*Growth_factor*EBITDA_margin, 2),
      "\nThe terminal value at this time given a multiple of:", EBITDA_Exit_Multiple,
      "and the above EBITDA for 2032, would equal:", round(Revenue_data[9]*Growth_factor*EBITDA_margin*EBITDA_Exit_Multiple, 2))  

```
